```python
import numpy as np
# vamos criar nosso primeiro array em numpy (reparem que temos o os numeros dentro de
# couchetes [] e usamos a função array do numpy. (np = numpy comumente usado)
temperaturas = np.array([22.1, 23.4, 21.0, 25.5, 24.3, 26.2, 27.0])

# Exibir o array de temperaturas
print("Temperaturas da semana:", temperaturas)

# Calculando a temperatura média, máxima e mínima
media = temperaturas.mean()
temp_max = temperaturas.max()
temp_min = temperaturas.min()

print(f"Temperatura média: {media:.2f}°C")
print(f"Temperatura máxima: {temp_max}°C")
print(f"Temperatura mínima: {temp_min}°C")

# Acessando a temperatura do quarto dia (Lembrete: indexação começa em 0)
temperatura_quarta = temperaturas[3]
print(f"Temperatura da quarta-feira: {temperatura_quarta}°C")

# Acessando as temperaturas do fim de semana (últimos dois dias)
temperaturas_fim_de_semana = temperaturas[-2:]
print(f"Temperaturas do fim de semana: {temperaturas_fim_de_semana}")

# Suponha que queremos aumentar todas as temperaturas em 2 graus Celsius
temperaturas_ajustadas = temperaturas + 2
print(f"Temperaturas ajustadas: {temperaturas_ajustadas}")
```
