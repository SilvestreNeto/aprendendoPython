```python
#exercício 01

n1 = float(input("Insira a primeira nota do aluno: "))
n2 = float(input("Insira a segunda nota do aluno: "))
media = (n1+n2)/2

print(f"\n A média do aluno foi: {media} \n")

if(media == 10):
    print("Aprovado com  distinção")
elif media >= 7 and media < 10:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
#exercício 03

salario = float(input("Insira o valor do salário: "))
percentual = 0.0

if salario > 0:
  if salario <= 1500:
    percentual = 0.2
  elif salario > 1500 and salario < 3000:
    percentual = 0.1
  else:
    percentual = 0.05

  salario_novo = salario + (salario*percentual)
  print(f"O salário antes era: R${salario:.2f}")
  print(f"O salário agora é: R${salario_novo:.2f}")
  print(f"O percentual de aumento foi de: {percentual*100:.0f}%")
  print(f"O aumento foi de: R${salario_novo-salario:.2f}")

else:
  print("O valor do salário não pode ser negativo")
```

```python
num_alunos = int(input("Insira o número de alunos: "))
nome_alunos = []
lista_de_notas = []
soma = 0

for j in range(num_alunos):
  nome = input(f"Insira o nome do aluno n° {j+1}: ")
  nome_alunos.append(nome)

for i in range(num_alunos):
  nota = float(input(f"Insira o valor da nota do aluno {nome_alunos[i]}: "))
  lista_de_notas.append(nota)
  soma = soma + nota

media = soma / num_alunos

print("\n")
print(f"A média das notas é: {media:.2f}\n")

for n in range(num_alunos):
  if lista_de_notas[n] >= media :
    print(f"O aluno {nome_alunos[n]} está ACIMA da média | NOTA: {lista_de_notas[n]}")
  else:
    print(f"O aluno {nome_alunos[n]} está ABAIXO da média | NOTA: {lista_de_notas[n]}")
```
