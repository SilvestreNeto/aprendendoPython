```python
#exercício 01
contador = 0
num = int(input("Insira um número"))

if num < 0:
    print("Número inválido")
else:
  while contador < num:
    contador += 1
    print(contador)
```

```python
#exercicio 02

notas = []
acimaDaMedia = []

while True:
    nota = float(input("Insira uma nota: "))
    if nota >= 0 and nota <= 10:
        notas.append(nota)
    if nota == -1:
      break

media = sum(notas) / len(notas)

for nota in notas:
    if nota > media:
        acimaDaMedia.append(nota)

print("======="*5)
print(f"Foram informadas {len(notas)} notas")
print(f"A média das notas é {media:.2f}")
print(f"Foram informadas {len(acimaDaMedia)} notas acima da média: {acimaDaMedia}")
print(f"Maior nota: {max(notas)}")
print(f"Menor nota: {min(notas)}")

```

```python
#Exercício 03

contador = 0
senha = 'bia2025'

while contador < 3:
  senha = input("Insira a senha: ")
  if senha == 'bia2025':
    print("Acesso permitido")
    break
  elif senha != 'bia2025':
    contador += 1
    print(f"Senha incorreta. Tentativa {contador} de 3.")
  else:
    print("Acesso negado. Entre em contato com o suporte")
```

```python
#Exercício 05

temperatura = []

for i in range(1,8):
  temperatura.append(float(input(f"Insira a temperatura do dia {i}: ")))

media = sum(temperatura) / len(temperatura)
print("========="*7)
print(f"A média das temperaturas é {media:.2f}")
print(f"A maior temperatura é {max(temperatura)}")
print(f"A menor temperatura é {min(temperatura)}")
```

```python
#Exercício 06 !!!!REVISAR!!!!
N = int(input("Digite um número inteiro positivo: "))

quantidade = 0

print("Números primos:")

for num in range(2, N + 1):
    eh_primo = True

    # Verifica divisores
    for i in range(2, num):
        if num % i == 0:
            eh_primo = False
            break

    if eh_primo:
        print(num, end=" ")
        quantidade += 1

print(f"\nQuantidade de números primos: {quantidade}")
```

```python
#Exercício 07
import random

num = random.randint(1, 100)
tentativas = 0

while True:
  valor = int(input("Insira um chute: "))
  tentativas += 1
  if (valor > num):
    print((f"O valor {valor} é muito alto \n"))
  elif (valor < num):
    print((f"O valor {valor} é muito baixo \n"))
  elif (valor == num):
    print(f"\nParabens, você acertou em {tentativas} tentativas")
    break
```

```python
#Exercício 08

nomesEquipamentos = []
valoresEquipamentos = []
numeroDeEquipamentos = int(input("Insira a quantidade de equipamentos: "))

for i in range(numeroDeEquipamentos):
  nome = input(f"Insira o nome do equipamento {i+1}: ")
  nomesEquipamentos.append(nome)

  valor = float(input(f"Insira o valor do equipamento {nome}: "))
  valoresEquipamentos.append(valor)

print("\nOs equipamentos são: ")
for i in range(numeroDeEquipamentos):
  print(f"{nomesEquipamentos[i]} | R$ {valoresEquipamentos[i]}")

print(f"\nValor total do inventário: R${sum(valoresEquipamentos)} ")

indiceMaisCaro = valoresEquipamentos.index(max(valoresEquipamentos))
indiceMaisBarato = valoresEquipamentos.index(min(valoresEquipamentos))

print(f"Equipamento mais caro: {nomesEquipamentos[indiceMaisCaro]} | valor: R${max(valoresEquipamentos)}")
print(f"Equipamento mais barato: {nomesEquipamentos[indiceMaisBarato]} | valor: R${min(valoresEquipamentos)}")
```
