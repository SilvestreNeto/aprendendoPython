```python
#exercício 01
contador = 0
num = int(input("Insira um número"))

if num < 0:
    print("Número inválido")
else:
  while contador < num:
    contador += 1
    print(contador)
```

```python
#exercicio 02

notas = []
acimaDaMedia = []

while True:
    nota = float(input("Insira uma nota: "))
    if nota >= 0 and nota <= 10:
        notas.append(nota)
    if nota == -1:
      break

media = sum(notas) / len(notas)

for nota in notas:
    if nota > media:
        acimaDaMedia.append(nota)

print("======="*5)
print(f"Foram informadas {len(notas)} notas")
print(f"A média das notas é {media:.2f}")
print(f"Foram informadas {len(acimaDaMedia)} notas acima da média: {acimaDaMedia}")
print(f"Maior nota: {max(notas)}")
print(f"Menor nota: {min(notas)}")

```

```python
#Exercício 03

contador = 0
senha = 'bia2025'

while contador < 3:
  senha = input("Insira a senha: ")
  if senha == 'bia2025':
    print("Acesso permitido")
    break
  elif senha != 'bia2025':
    contador += 1
    print(f"Senha incorreta. Tentativa {contador} de 3.")
  else:
    print("Acesso negado. Entre em contato com o suporte")
```
